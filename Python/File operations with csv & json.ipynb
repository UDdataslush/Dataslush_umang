```python
import json

# JSON string
json_string = '{"name": "Alice", "age": 25, "is_student": false}'

# Parse JSON string into a Python dictionary
data = json.loads(json_string)

print(data)          # Output: {'name': 'Alice', 'age': 25, 'is_student': False}
print(data['name'])  # Output: Alice




# Assuming a file named 'data.json' exists
with open('industry.json', 'r') as file:
    data = json.load(file)

print(data)



# Python dictionary
data = {
    "name": "Bob",
    "age": 30,
    "is_student": True
}

# Write to a file
with open('output.json', 'w') as file:
    json.dump(data, file, indent=4)  # Pretty-print with indentation




# Python dictionary
data = {
    "name": "Charlie",
    "age": 35,
    "is_student": False
}

# Convert to JSON string
json_string = json.dumps(data, indent=4)
print(json_string)




# Nested data
data = {
    "students": [
        {"name": "Alice", "age": 25},
        {"name": "Bob", "age": 30}
    ],
    "class": "Physics"
}

# Convert to JSON string
json_string = json.dumps(data, indent=4)
print(json_string)

# Accessing nested data
print(data['students'][0]['name'])  # Output: Alice





try:
    with open('industry.json', 'r') as file:
        data = json.load(file)
    print(data)
except json.JSONDecodeError as e:
    print(f"Error decoding JSON: {e}")



class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

# Custom encoder
def person_to_dict(obj):
    if isinstance(obj, Person):
        return {"name": obj.name, "age": obj.age}
    raise TypeError("Type not serializable")

# Serialize
person = Person("Daisy", 28)
json_string = json.dumps(person, default=person_to_dict, indent=4)
print(json_string)


```
