```python
import pandas as pd
import numpy as np

# Read the CSV file
df = pd.read_csv(r'C:\Users\Admin\Desktop\task1\python\Titanic-Dataset.csv')
print(df)

df.describe()
```

```python
# create

df['FamilySize'] = df['SibSp'] + df['Parch']
df['FarePerPerson'] = df['Fare'] / (df['FamilySize'] + 1)
```

```python
# update

df['Age'] = df['Age'].fillna(df['Age'].median())  # Fill missing ages
 
df.loc[df['Age'] < 18, 'Category'] = 'Child'       # Add category column
```

```python
# delete

df = df.drop(columns=['Cabin', 'Ticket'])          # Remove unnecessary columns
```

```python
# Common Methos Using in python
```

```python
#1 Sorting Method

sorted_df = df.sort_values(by='Age')  # Sort by Age in ascending order
sorted_df = df.sort_values(by='Age', ascending=False)  # Sort by Age in descending order

sorted_df = df.sort_values(by=['Pclass', 'Age'])
sorted_df = df.sort_values(by=['Pclass', 'Age'], ascending=[True, False])
```

```python
#2 Filtering Method
#Age above 30 and Fare greater than 50:

filtered_df = df[(df['Age'] > 30) & (df['Fare'] > 50)]
filtered_df = df[(df['Age'] < 18) | (df['Survived'] == 1)]

filtered_df = df[df['Embarked'] == 'S']

```

```python
filtered_df = df[(df['Age'] > 30) & (df['Survived'] == 1)]
sorted_filtered_df = filtered_df.sort_values(by='Fare', ascending=False)
print(sorted_filtered_df.head())
```

```python
# group by

average_age_by_class = df.groupby('Pclass')['Age'].mean()
print(average_age_by_class)


```

```python
survival_rate = df.groupby(['Pclass', 'Sex'])['Survived'].mean()
print(survival_rate)


```

```python
grouped_stats = df.groupby('Pclass').agg({
    'Age': ['mean', 'median'],
    'Fare': ['mean', 'max']
})
print(grouped_stats)



```

```python
grouped = df.groupby('Pclass')
filtered = grouped.filter(lambda x: len(x) > 100)  # Groups with more than 100 passengers
print(filtered)
```

```python
# Numpy

#1 reshape


# Extract 'Age' column as a NumPy array
age_array = df['Age'].dropna().values  # Drop NaNs to avoid issues with reshaping

# Reshape to 2D (e.g., 2 rows and n columns)
reshaped_age = age_array.reshape(2, -1)  # -1 auto-calculates the number of columns
print(reshaped_age)
```

```python
# aggregation

fare_array = df['Fare'].values

# Aggregation operations
print("Mean Fare:", np.mean(fare_array))
print("Max Fare:", np.max(fare_array))
print("Min Fare:", np.min(fare_array))
print("Standard Deviation:", np.std(fare_array))
```

```python
# indexing
age_array = df['Age'].fillna(0).values  # Replace NaN with 0 for indexing
print("First age:", age_array[0])
print("First 5 ages:", age_array[:5])

# Accessing specific rows and columns in a 2D array
fare_pclass = df[['Fare', 'Pclass']].values
print("First row:", fare_pclass[0])        # First row
print("First column (Fare):", fare_pclass[:, 0])  # All rows of the first column
```

```python
#slicing

fare_pclass = df[['Fare', 'Pclass']].values

# Extract first 3 rows
print("First 3 rows:\n", fare_pclass[:3])

# Extract first column (all rows)
print("First column (Fare):\n", fare_pclass[:, 0])

# Extract rows 2-4 and columns 1-2
print("Subarray:\n", fare_pclass[1:4, 0:2])
```
